## GDP vs Alcohol

It appears that weathy countries are more likely, but not guarenteed, to have high alcohol consumption rates.
In fact, it appears that weathy countries outside the Middle East almost always have high or moderate rates.

```python
import pandas as pd
gdp = pd.read_csv("gdp.csv", names=["Country", "GDP"], header=0).set_index("Country")
alco = pd.read_csv("booze.csv", names=["Country", "Liters"], header=0).set_index("Country")
gdprating = pd.qcut(gdp['GDP'], 3, labels=("Poor", "Average", "Rich"))
alcorating = pd.qcut(alco['Liters'], 5, labels=("Low", "Medium-Low", "Medium", "Medium-High", "High"))
ratings = pd.merge(gdprating.to_frame(), alcorating.to_frame(), left_index=True, right_index=True)
print(ratings.head())
print(ratings.tail())
```

**A plot of alcohol consumption over a log scale of gdp shows a very loose correlation.**

```python
import matplotlib.pyplot as plt
import numpy as np
combo = pd.merge(gdp, alco, on="Country")
plt.xscale('log')
plt.scatter(combo["GDP"], combo["Liters"])
```
